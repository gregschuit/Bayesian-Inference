```python
import numpy as np
from numpy.random import beta as Beta
from matplotlib import pyplot as plt
from scipy import stats
from random import random
```

```python
# USAR LOGARITMOS DE LAS PROBABILIDADES DESPUES DE ESTA ACTIVIDAD, AHORA NO ES NECESARIO.
```

```python
# DATOS
theta = 0.4  # probabilidad de cara
N = 10
data = []
for i in range(10):
    data.append(1 if random() < theta else 0)
data
```

```python
Nc = len([d for d in data if d == 1])  # Numero de Caras
Ns = len([d for d in data if d == 0])  # Numero de Sellos
Nc, Ns, Nc + Ns
```

```python
# Conteos imaginarios a priori
alpha = 2
beta = 2
```

```python
x = np.linspace(0, 1, 100)
plt.figure(figsize=(15, 6))

# Likelihood normalizada es una beta
plt.plot(x, stats.beta.pdf(x, Nc + 1, Ns + 1), color='blue')

# Prior es una beta
plt.plot(x, stats.beta.pdf(x, alpha, beta), color='red')

# A posterior
plt.plot(x, stats.beta.pdf(x, Nc + alpha, Ns + beta), color='purple')

print(np.mean(stats.beta.pdf(x, Nc + alpha, Ns + beta)))

plt.show()
```
