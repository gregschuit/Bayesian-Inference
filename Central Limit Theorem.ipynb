# Central Limit Theorem

```python
import numpy as np
from matplotlib import pyplot as plt
from random import sample
```

```python
np.random.seed(123)
```

### Uniform Distribution Sampling

```python
random_data = list(np.random.uniform(0, 1, 2000))
np.mean(random_data)
```

```python
plt.figure(figsize=(15, 6))

sizes = [i for i in range(10, 2000, 10)]
plt.scatter(sizes, [np.mean(sample(random_data, k)) for k in sizes])

x = np.linspace(10, 2000, 100)
plt.plot(x, np.mean(random_data)*np.ones(len(x)), color='red')

plt.show()
```

### Multinomial Distribution: Dice Rolls

```python
np.random.seed(2342)
```

```python
plt.figure(figsize=(16, 6))
plt.ylim(1, 6)

sizes = [k for k in range(2, 400, 2)]

for k in sizes:
    dice_rolls = np.random.multinomial(k, [1/6]*6)
    loaded_rolls = np.random.multinomial(k, [1/8]*5 + [3/8])  # dado cargado  
    plt.scatter(k, np.dot(dice_rolls, np.array([1,2,3,4,5,6])) / k, color='red')
    plt.scatter(k, np.dot(loaded_rolls, np.array([1,2,3,4,5,6])) / k, color='green')

plt.plot([2, 400], [3.5]*2)  # Esperanza teorica

plt.show()
```

### Gaussian Distribution Analisis

```python
mean = np.array([2,6])
```

```python
precov = np.array([[-6, 5],
                   [5, 2]])
cov = np.dot(precov, precov.T)
cov
```

```python
x, y = np.random.multivariate_normal(mean, cov, 3000).T


plt.scatter(x, y)
plt.show()
```

```python
plt.hist([y[i] for i in range(len(y)) if -20 < x[i] < -15], bins=10)
plt.show()
plt.hist([y[i] for i in range(len(y)) if -15 < x[i] < -10], bins=10)
plt.show()
plt.hist([y[i] for i in range(len(y)) if -10 < x[i] < -5], bins=10)
plt.show()
plt.hist([y[i] for i in range(len(y)) if -5 < x[i] < 0], bins=10)
plt.show()
plt.hist([y[i] for i in range(len(y)) if 0 < x[i] < 5], bins=10)
plt.show()
plt.hist([y[i] for i in range(len(y)) if 5 < x[i] < 10], bins=10)
plt.show()
plt.hist([y[i] for i in range(len(y)) if 10 < x[i] < 15], bins=10)
plt.show()
plt.hist([y[i] for i in range(len(y)) if 15 < x[i] < 20], bins=10)
plt.show()
```
